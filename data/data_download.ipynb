## Data Download

```python
import pandas as pd
import urllib2
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import os
import requests
```

```python
base = "http://www.tennis-data.co.uk/"
years = range(2001,2018)
end = "/usopen.csv"


for year in years:
    url = base+str(year)+end
    response = urllib2.urlopen(url)
    html = response.read()

    with open(str(year)+'.csv', 'wb') as f:
        f.write(html)
```
